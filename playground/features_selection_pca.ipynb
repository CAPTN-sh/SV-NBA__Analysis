# 1. <a id='toc1_'></a>[Features Selection](#toc0_)

This notebook contains the code to select the most important features for the model. 

Here filter mothod is used to select the features. The features are selected using the `feature_importance_` attribute of the model. The features are then ranked and the top 10 features are selected. The selected features are then used to train the model.

Filter-based feature selection methods for unsupervised data typically rely on statistical measures or heuristic approaches to rank features based on their intrinsic characteristics rather than on a specific learning algorithm. Here are a few filter-based methods along with how you might associate each selected feature with its importance:

+ Variance Threshold</br>
Compute the variance of each feature. Features with low variance are less informative and can be removed.
Associate the variance value directly as the feature importance.
Correlation Coefficient:

+ Calculate the correlation coefficient between each pair of features.</br>
  Features highly correlated with other features might contain redundant information. You can select one of each highly correlated pair or remove one randomly.
Associate the absolute value of the correlation coefficient as the feature importance.
Mutual Information:

+ Measure the mutual information between each feature and the cluster labels.</br>
Features with high mutual information are more informative for clustering.
Associate the mutual information value as the feature importance.
Distance-based Methods:

+ Compute the distance between instances in the feature space and analyze the distribution of distances.</br>
Features that contribute to larger distances between instances might be more important for clustering.
Associate the distance measure (e.g., mean distance or median distance) as the feature importance.

+ ``sklearn.feature_selection`` module is used for feature selection/dimensionality reduction.
+ Goal:
  + Improve estimators accuracy scores
  + Avoiding overfitting
  + Reduce the computational cost
  + Improve the comprehensibility of the model
+ There are three main strategies:
  + Univariate statistics: Select the best features based on univariate statistical tests
  + Model-based selection: Use a supervised model to judge the importance of each feature
  + Iterative selection: Build a model on initial features and then iteratively remove the least important feature
+ Feature selection methods can also be categorised into:
  + Filter methods: Select features based on their scores in various statistical tests
  + Wrapper methods: Select features based on the performance of a model trained with the selected features
  + Embedded methods: Select features based on the importance of their contribution to the model
+ Feature selection can be done in four ways:
  + **SelectKBest**: Select features according to the k highest scores
  + **SelectPercentile**: Select features according to a percentile of the highest scores
  + **SelectFpr**: Select features based on a false positive rate test
  + **SelectFdr**: Select features based on an estimated false discovery rate
  + **SelectFwe**: Select features based on family-wise error

### PCA

while PCA is not explicitly designed for feature selection, it can be used indirectly for this purpose by considering the importance of features based on the variance they explain in the data.

Principal Component Analysis (PCA) is primarily a dimensionality reduction technique, but it can also be used for feature selection indirectly. Here's how PCA can be used for feature selection:

1. **Dimensionality Reduction**: PCA reduces the dimensionality of the data by transforming the original features into a new set of orthogonal (uncorrelated) features called principal components. These principal components are ordered in such a way that the first component explains the maximum variance in the data, the second component explains the maximum variance remaining after the first, and so on.

2. **Variance Retention**: When performing PCA, each principal component captures a certain amount of variance in the data. By selecting a subset of the principal components, you can retain a certain percentage of the total variance while reducing the dimensionality of the data. 

3. **Feature Importance**: In PCA, the importance of a feature can be inferred from the amount of variance it explains in the data. Features that contribute more to the variance will have higher importance. This is because these features contain more information about the underlying structure of the data.

4. **Selection of Principal Components**: By selecting the top N principal components that capture the most variance in the data, you effectively select the corresponding original features that contribute the most to this variance. These selected principal components, and hence the original features, are considered important for representing the dataset.

5. **Dimensionality Reduction vs. Feature Selection**: Although PCA primarily focuses on dimensionality reduction, by selecting a subset of principal components, you implicitly select a subset of original features. This subset contains the features that are most relevant for explaining the variance in the data, effectively serving as a form of feature selection.

+ Find the best threshold value for the variance threshold method using Optuna. Using the silhouette score with k-means clustering.
  > **NOTE**:</br> Assuming that the number of clusters is $10$.
    

```python
def feature_selection_pca(df: pd.DataFrame, n_components: int) -> Tuple[PCA, pd.DataFrame]:
    """
    Perform feature selection using PCA.

    Args:
        df (DataFrame): The input dataframe containing the features.
        n_components (int): The number of principal components to keep.

    Returns:
        tuple: A tuple containing two objects:
            - pca_model (PCA): The trained PCA model.
            - fs_df (DataFrame): DataFrame containing selected features and their importance scores.
    """
    
    # Instantiate a place holder for the variance threshold method (vtm) selected features >
    fs_df = pd.DataFrame(columns=['selected_features', 'feature_importance'])
    
    # Instantiate PCA >
    pca = PCA(n_components=n_components)
    
    # Standardized data >
    # scaler = StandardScaler()
    # df_standardized = scaler.fit_transform(df)
    
    # Fit the PCA model >
    pca.fit(df)  #(df_standardized)
    
    # Extract feature importance scores >
    feature_scores = pca.explained_variance_ratio_
    
    # Normalize reconstruction errors using Normalizer
    normalizer = Normalizer()
    normalized_scores = normalizer.fit_transform(feature_scores)
    
    # Sort features based on importance scores >
    sorted_features = sorted(zip(df.columns, normalized_scores), key=lambda x: x[1], reverse=True)
    
    # Normalize the reconstruction errors to get scores that represent the importance of each feature. 
    selected_features = [feat for feat, _ in sorted_features]
    feature_scores = [score for _, score in sorted_features]
    
    # Create DataFrame to store selected features and scores
    fs_df['selected_features'] = selected_features
    fs_df['feature_importance'] = feature_scores
    
    # fs_df = pd.DataFrame(sorted_features, columns=['selected_features', 'feature_importance'])
    
    return pca, fs_df
```

```python
def optimise_pca_hyperparameters(df: pd.DataFrame, n_components_range: Tuple[int, int]) -> int:
    """
    Optimize the number of principal components for PCA using GridSearchCV.

    Args:
        df (DataFrame): The input DataFrame for training the PCA model.
        n_components_range (tuple): Range of the number of principal components to explore.

    Returns:
        int: The best number of principal components found by GridSearchCV.
    """
    param_grid = {'pca__n_components': list(range(n_components_range[0], n_components_range[1]+1))}
    
    # Constructing the pipeline
    pca = PCA()
    n_samples = df.shape[0]
    kmeans = KMeans(n_clusters=min(30, n_samples//2), random_state=42)
    pipeline = Pipeline([('pca', pca), ('kmeans', kmeans)])
    
    # Define a function to compute silhouette score
    def silhouette_scorer(estimator, X):
            labels = estimator.predict(X)
            return silhouette_score(X, labels)
                    
    grid_search = GridSearchCV(estimator=pipeline, 
                               param_grid=param_grid, 
                               cv=None, 
                               scoring=silhouette_scorer,  # 'neg_mean_squared_error',
                               n_jobs=n_jobs,
                               verbose=3)
    grid_search.fit(df)
    
    best_n_components = grid_search.best_params_['pca__n_components']
    
    return best_n_components
```

```python
%%time
n_components_range = (2, df.shape[1])  #min(len(df.columns), 10))
best_n_components = optimise_pca_hyperparameters(df, n_components_range)

print(f"Best number of principal components: {best_n_components}")
```

```python
_, fs_df = feature_selection_pca(df, best_n_components)

# Free up memory
del df
```
